# Python Operators

Python operators are the symbols or keywords that Python uses to perform mathematical, comparison, and logical operations.

Python operators act on data (numbers, strings, variables) to produce results.

```python
a = 10
b = 3

print(a + b)      # Arithmetic operator (+)
print(a > b)      # Comparison operator (>)
print(a and b)    # Logical operator (and)
```

**Python has different types of operators:**

-Arithmetic operators → +, -, *, /, %

-Comparison operators → ==, !=, >, <

-Logical operators → and, or, not

-Assignment operators → =, +=, -=

-Bitwise operators → &, |, ^, <<, >>

-Identity operators → is, is not

-Membership operators → in, not in

# Comparison Operator


Used to compare two values. Result is always True or False.

**Operators:**

(==)Equal to

(!=)Not equal to

(>)Greater than

(<)Less than

(>=)Greater than or equal to

(<=)Less than or equal to

```python
x,y=7,3
print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
```

```python
a = 25
b = 12
c = 40
if a > b and a > c:
    print("A is greatest")
elif b > a and b > c:
    print("B is greatest")
else:
    print("C is greatest")
```

```python
marks = 37
if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
n = 14
if n % 2 == 0 and n > 0:
    print("Even and Positive")
else:
    print("Condition not satisfied")
```

```python
age = 45
member = True
if age > 60 or member:
    print("Discount Available")
else:
    print("No Discount")
```

# Logical Operator


Used to combine multiple conditions.

**Operators:**

and → True if both conditions are True

or → True if at least one condition is True

not → Reverses the condition

```python
p,q=True,False
print(p and q) #All elements should be true to get true
print(p or q) #atleasts one of the element should be true
print(not p) 
```

```python
username = "admin"
password = 1234
if username == "admin" and password == 1234:
    print("Login Successful")
else:
    print("Login Failed")
```

```python
#Nested logical operator
x = 25
result = (x > 10 and x < 50) or (x == 100)
print(result)
```

```python
#Floor Division
print(15 // 4)
print(29 // 6)
print(100 // 9)
```
