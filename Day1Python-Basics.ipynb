# Python Basics

**What is Python?**

Python is a high-level, general-purpose programming language known for being simple, readable, and powerful.

**Key features:**

-Easy to read and write

-Interpreted language (runs line-by-line)

-Portable (runs on Windows, Mac, Linux)

-Object-oriented, functional & procedural support

-Large community and libraries

-Used in web development, AI, machine learning, automation, data science, gaming, etc.

**Why Use Python?**

Python is popular because:

You can do a lot with very little code

**Huge library support:**

NumPy, Pandas, Matplotlib → Data Science

TensorFlow, PyTorch → AI/ML

Django, Flask → Web development

Selenium → Automation

Great for beginners and professionals

# Data Types

**What is a Data Type?**

A data type defines the kind of data a variable can store and what operations can be performed on it.

Categories of Data Types:-

**1. Primitive Data Types**

**2. Non‑Primitive (Reference) Data Types**

**Primitive Data Types**

1. **Integer (int)**

Integers are whole numbers without decimals.
They can be positive, negative, or zero.

```python
x = 10
y = -5
z = 0
print(type(x))   # <class 'int'>
```

```python
x = 5
y = 10

x, y = y, x
print("x =", x, "y =", y)
```

2. **Float (float)**

Float means decimal numbers.

```python
pi = 3.14
marks = 89.5
print(type(pi))   # <class 'float'>
```

```python
x = 5.6
y = 8.34
print("Sum =", x + y)
```

```python
r = 4.5
pi = 3.14159
area = pi * r * r
print("Area of circle =", area)
```

3. **Boolean (bool)**

Boolean represents True or False values.

```python
is_active = True
is_login = False
print(type(is_active))   # <class 'bool'>
```

```python
age = 20
is_adult = age > 18
print("Adult:", is_adult)
```

4. **String (str)**

Strings represent text, enclosed in single ' ' or double " " quotes.

```python
name = "Rahul"
msg = 'Hello World'
print(type(name))   # <class 'str'>
```

```python
text = "Data Types"
print("Length =", len(text))
```
