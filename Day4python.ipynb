# LIST

A list is an ordered, changeable (mutable) collection of items.

Written using square brackets [ ].

**Characteristics:**

-Mutable → elements can be added, removed, or changed

-Allows duplicates

-Supports different data types

-Ordered (index starts from 0)

# Common List Methods

append() – add at end

insert() – add at specific index

remove() – remove by value

pop() – remove by index

sort() – sort list

reverse() – reverse list

extend() – add another list

count() – count occurrences

index() – find index

# List Operations

Concatenation: list1 + list2

Repetition: list1 * 3

Membership: "a" in lst

# Mutability

Lists can be modified.

**Use List when:**

-You need to modify data

-Add or delete items frequently

# TUPLE

A tuple is an ordered, unchangeable (immutable) collection of items.

Written using parentheses ( ).

# Characteristics

Immutable → cannot change elements after creation

Faster than lists

Allows duplicates

Can contain different data types

Can be used as dictionary keys

# Tuple Functions

count() – count occurrences

index() – find index

len() – number of elements

max() / min() – largest & smallest

# Tuple Operations

Concatenation: (1,2) + (3,4)

Repetition: (5,) * 3

Membership: 10 in t

# Immutability

Tuple cannot be modified.

**Use Tuple when:**

-Data should NOT change

-Faster performance needed

-Use as dictionary keys

```python
tup1=()
type(tup1)
```

```python
tup2=(10,24,30,5,7) #tuple of numbers or integer
tup3=(2.0,4.7,7.8,78.6) #tuple of float
tup4=('marvel','DCEU','disney') #tuple of strings
tup5=(37,3.76,'Data') #tuple of mixed datatype
tup6=(1,3,(12,34,57,),(10,20,4.0),('Alpha','Beta','Charlie'))
```

```python
#Indexing
tup2[1]
```

```python
print(tup6[4][0])
print(tup4[0])
```

```python
#Slicing
print(tup2[2:])
```

```python
tup6=('mon','tue','wed','thur','fri','sat','sun')
print(tup6[5:]) #weekend slicing
```

```python
#Count Method
tup9=('one','one','two','three','four','two','two')
tup9.count('one')
tup9.index('two') #if we have more than times of index no. it will consider only first index no.
```

```python
#Membership
print('three' in tup9)
print('five' in tup9) 
```

```python
#Sorting
tup10=(8,7,8,3,2,6,2,1,4)
print(sorted(tup10))
print(sorted(tup10,reverse=True)) #by false it will give assending order of tuple,by default.
```
