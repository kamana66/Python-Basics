**String Indexing**

Indexing gives access to individual characters.
Index starts from 0.

```python
s = "python"
s[0]     # 'p'
s[3]     # 'h'
s[-1]    # 'n' (last letter)
```

```python
#string indexing
print(str[0])
print(str[-1])
```

```python
name="Python"       #STRING=sequences of characters inside quotes.
print(name[0])      #INDEXING
print(name[-1])     #REVERSE INDEXING
print(name[0:3])    #SLICING
```

**String Concatenation**

Joining two or more strings using + or join().

```python
s1 = "Hello"
s2 = "World"

s1 + " " + s2      # 'Hello World'
"".join([s1, s2])  # 'HelloWorld'
```

```python
#string concatenation
s1="DATA"
s2="SCIENCE"
s3=s1+s2
print(s3)
print(s1+" "+s2)
```

**String Membership Operators**

Used to check whether a character or substring exists inside a string.

**Operators:**

in → returns True if substring is found

not in → returns True if substring is not found

```python
#string membership
str="HELLO PYTHON"
print("HELLO" in str)
print("science" in str)
```

**String Partitioning**

partition()

**Splits a string into three parts:**

1.before separator,

2.the separator itself,

3.after separator.

```python
#string partitioning
string="This is separator of the line"
print(string.partition("separator"))
```

**String Slicing**

```python
s = "python"
s[1:4]     # 'yth'
s[:3]      # 'pyt'
s[3:]      # 'hon'
s[::-1]    # 'nohtyp'  (reverse)
```

```python
#string slicing
str1="HELLO PYTHON"
print(str1[0:4])
print(str1[6:])
print(str1[-6:])
```

**Strip Functions**
  
strip()-Removes spaces (or characters) from both ends.

lstrip()-Removes from left side.

rstrip()-Removes from right side.

```python
s = "   python   "
s.strip()   # 'python'
s.lstrip()  # 'python   '
s.rstrip()  # '   python'
```

```python
#strip function
text2=" HELLO "
(text2.strip())
```

```python
(text2.lstrip())
```

```python
(text2.rstrip())
```

```python
text3="***HELLO***"
print(text3.strip("*"))
```

**Uppercase / Lowercase Methods**

upper()-Converts all characters to uppercase.

lower()-Converts to lowercase.

title()-Capitalizes first letter of every word.

capitalize()-Capitalizes only the first character.

```python
s = "hello world"
s.upper()        # 'HELLO WORLD'
s.lower()        # 'hello world'
s.title()        # 'Hello World'
s.capitalize()   # 'Hello world'
```

```python
text="kantara"
print(text.upper()) #to convert string to uppercase
print(text.lower()) #to convert string to lowercase
```

**startswith() Method**

Checks if a string begins with a specific substring.
    
**endswith() Method**

Checks if a string ends with a specific substring.

```python
s = "programming"
s.startswith("pro")    # True
s.startswith("gram")   # False
```

```python
s = "programming"
s.endswith("ing")      # True
s.endswith("gram")     # False
```

```python
text5="Hello Python"
print(text5.startswith("Hello")) #startswith function
print(text5.endswith("Python")) #endswith function
```
